```python
# sampling to 10% of data from 42k rows which will be complex for knn algorithm
df=df.sample(frac=0.1,random_state=42)
```

```python
# find unqiue class labels
df['label'].value_counts()
```

```python
#prepare x and y
x=df.drop(columns=['label'])
y=df['label']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/4,random_state=42)
```

```python
# percentage=len(x_train)/len(df)
# percentage
```

```python
from sklearn.neighbors import KNeighborsClassifier

# as we already have the x_train and y_train data 
# we now train knn model
knn=KNeighborsClassifier(n_neighbors=3,metric="minkowski")
knn.fit(x_train,y_train)
#now predict 
predicted=knn.predict(x_test)
print(predicted)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate confusion matrix
cm = confusion_matrix(y_test, predicted)

# Visualizing the confusion matrix using seaborn
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
